# 作業 : (Kaggle)房價預測

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

import warnings
warnings.filterwarnings("ignore")
```

```python
data_path = 'data_part02/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
import itertools
low = range(300, 700, 100)
high = range(2000, 4000, 500)

value_range = list(itertools.product(low, high))
value_range
```

```python
print(len(value_range))
print(value_range[1])
print(value_range[1][0])
print(value_range[1][1])
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
df1 = df.copy()

fig = plt.figure(figsize = (25,25))

for i in range(16):
    
    vrange = value_range[i]
    low = value_range[i][0]
    high = value_range[i][1]
    
    # 調整上下界
    df1['1stF1rSF'] = df1['1stFlrSF'].clip(low, high)

    # 做線性迴歸, 觀察分數
    train_X = MMEncoder.fit_transform(df1)
    estimator = LinearRegression()
    cvscore_mean = cross_val_score(estimator, train_X, train_Y, cv=5).mean()
    
    titlename =  str(vrange) + " CV score mean: " + str(round(cvscore_mean, 6))
    
    # plots
    ax = fig.add_subplot(4, 4, i+1)
    sns.regplot(x = df1['1stF1rSF'][:train_num], y = train_Y, ax = ax)
    plt.title(titlename)
plt.show()
```

```python
# 16種組合中 最好的 (300, 2500)
df1['1stF1rSF'] = df1['1stFlrSF'].clip(300, 2500)
sns.regplot(x = df1['1stF1rSF'][:train_num], y = train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df1)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
plow = [0.01, 0.05, 0.1]
phigh = [0.99, 0.95, 0.9]

prange = list(zip(plow, phigh))
print(prange)
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
fig = plt.figure(figsize = (30,10))

for i in range(3):
    
    pt = prange[i]
    low = np.percentile(df['1stFlrSF'], pt[0]*100)
    high = np.percentile(df['1stFlrSF'], pt[1]*100)

    # 移除離群值
    df2 = df[(df['1stFlrSF'] > low) & (df['1stFlrSF'] < high)]
    train_Y = train_Y[(df['1stFlrSF'] > low) & (df['1stFlrSF'] < high)]
    
    # 做線性迴歸, 觀察分數
    train_X = MMEncoder.fit_transform(df2)
    estimator = LinearRegression()
    cvscore_mean = cross_val_score(estimator, train_X, train_Y, cv=5).mean()
    
    titlename =  str(pt) + " CV score mean: " + str(round(cvscore_mean, 6))
    
    # plots
    ax = fig.add_subplot(1, 3, i+1)
    sns.regplot(x = df2['1stFlrSF'], y = train_Y, ax = ax)
    plt.title(titlename)
plt.show()
```

###  以上述測試狀況, 排除離群值頭尾0.01% 樣本較好
